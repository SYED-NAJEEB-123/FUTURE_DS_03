```python
import pandas as pd

df = pd.read_csv('../data/bank-full.csv', sep=';')
df.head()

```

```python
df['conversion'] = df['y'].map({'yes':1, 'no':0})
df['conversion'].value_counts()
```

```python
conversion_rate = df['conversion'].mean()*100
print("Overall Conversion Rate:", round(conversion_rate,2), "%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(x='conversion', data=df)
plt.title("Converted vs Not Converted")
plt.figure(figsize=(6,4))
sns.countplot(x='conversion', data=df)
plt.title("Converted vs Not Converted")
plt.savefig("../images/conversion_distribution.png", bbox_inches='tight')
plt.close()

```

```python
plt.figure(figsize=(7,4))
sns.histplot(data=df, x='duration', hue='conversion', bins=50)
plt.title("Call Duration vs Conversion")
plt.figure(figsize=(7,4))
sns.histplot(data=df, x='duration', hue='conversion', bins=50)
plt.title("Call Duration vs Conversion")
plt.savefig("../images/call_duration_conversion.png", bbox_inches='tight')
plt.close()

```

```python
plt.figure(figsize=(7,4))
sns.boxplot(x='conversion', y='campaign', data=df)
plt.title("Number of Contacts vs Conversion")
plt.figure(figsize=(7,4))
sns.boxplot(x='conversion', y='campaign', data=df)
plt.title("Number of Contacts vs Conversion")
plt.savefig("../images/contact_conversion.png", bbox_inches='tight')
plt.close()

```

```python
## Key Insights

1. Overall conversion rate is approximately 1112%, indicating low campaign success rate.
2. Customers with longer call duration show significantly higher conversion probability.
3. Repeated contact attempts reduce the likelihood of subscription.
4. Most successful conversions occur within the first few interactions.

## Business Recommendations

 Improve targeting to contact more relevant customers
 Encourage agents to have meaningful conversations
 Limit repeated calls to avoid customer fatigue
 Focus on engagement rather than call volume
```
